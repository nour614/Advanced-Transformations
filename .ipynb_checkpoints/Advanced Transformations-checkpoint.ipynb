## Info

```python
info_df[["Hero", "Publisher"]] = info_df["Hero|Publisher"].str.split('|',expand=True)
info_df.head()
```

```python
# drop original 'Hero|Publisher' column
info_df = info_df.drop(columns = ['Hero|Publisher'])


info_df.head()
```

```python
# convert strings in dictionary to using double instead of single quotes 
info_df['Measurements'] = info_df['Measurements'].str.replace("'", '"')


measurement = info_df.loc[0, 'Measurements']
print(measurement)
```

```python
# use .apply with json.loads to convert strings to dicts
info_df['Measurements'] = info_df['Measurements'].apply(json.loads)


measurement = info_df.loc[0, 'Measurements']
print(type(measurement))
```

```python
# unpack 'Measurements' column (dictionaries) into separate
# columns for each dictionary item
height_weight = info_df['Measurements'].apply(pd.Series)


height_weight
```

```python
# concatenate new columns to original df
info_df = pd.concat((info_df, height_weight), axis = 1)


info_df.head()
```

```python
# drop original column
info_df = info_df.drop(columns = ['Measurements'])


info_df.head()
```

```python
# change 'Height' and 'Weight' column names
info_df.rename(columns = {'Height': 'Height (cm)',
                      'Weight': 'Weight (kg)'}, 
            inplace = True)


info_df.head()
```

```python
# take off 'cm' and 'kg' from cell values in 'Height' and 'Weight'
to_replace = [' cm', ' kg']

for char in to_replace:
    info_df['Height (cm)'] = info_df['Height (cm)'].str.replace(char, 
                                                          '', 
                                                          regex = False)
    info_df['Weight (kg)'] = info_df['Weight (kg)'].str.replace(char, 
                                                          '', 
                                                          regex = False)
    

info_df[['Height (cm)', 'Weight (kg)']].head()
```

```python
# convert 'Height (cm)' and 'Weight (kg)'columns from object-types to numeric

info_df['Height (cm)'] = info_df['Height (cm)'].astype(float)
info_df['Weight (kg)'] = info_df['Weight (kg)'].astype(float)


info_df.info()
```

## Powers

```python
# check type of powers['Powers'] column
power = powers_df.loc[0, 'Powers']
print(type(power))
print(power)
```

```python
# split string on comma
powers_df['powers_split'] = powers_df['Powers'].str.split(',')


powers_df[['Powers', 'powers_split']].head()
```

```python
exploded = powers_df.explode('powers_split')
exploded[['hero_names', 'Powers', 'powers_split']].head()
```

```python
# save unique values from exploded column 
new_columns= exploded['powers_split'].dropna().unique()

new_columns
```

```python
# create new column for each superpower in the list


subsets = [powers_df]

for col in new_columns:
    subset = pd.DataFrame()
    subset[col] = powers_df['Powers'].str.contains(col)
    subsets.append(subset)
    all_powers = pd.concat(subsets, axis = 1)
    

all_powers.head()
```

```python
# drop 'Powers' and 'powers_split' columns
all_powers.drop(columns = ['Powers', 'powers_split'], inplace = True)


all_powers.head()
```

```python
all_powers.rename(columns = {'hero_names': 'Hero'}, 
                     inplace = True)


all_powers.head()
```

```python
all_powers.info()
```

```python
info_df.info()
```

```python
# combine two dataframes on info_df and all_powers
super_heros = pd.merge(info_df, 
                   all_powers, 
                   on = 'Hero')


super_heros.head()
```

```python
super_heros.info()
```
